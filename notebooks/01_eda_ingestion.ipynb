```python
import pandas as pd

# lê só 100k linhas para exploração
df = pd.read_csv("data/raw/train_ver2.csv", nrows=100_000, low_memory=False)

print(df.shape)
df.head(3)
```

```python
df.dtypes
```

```python
for col in df.select_dtypes(include="object").columns:
    max_len = df[col].astype(str).str.len().max()
    uniques = df[col].dropna().unique()[:8]
    print(f"{col:35} | max_len: {max_len:4} | sample: {uniques}")
```

```python
null_pct = (df.isnull().sum() / len(df) * 100).round(1)
null_pct[null_pct > 0].sort_values(ascending=False)
```

```python
product_cols = [c for c in df.columns if c.startswith("ind_") and c.endswith("ult1")]
for col in product_cols:
    print(f"{col:30} | unique: {df[col].dropna().unique()}")
```
