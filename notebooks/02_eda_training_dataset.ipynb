```python
import pandas as pd
import os

print("Current dir:", os.getcwd())

df = pd.read_parquet("../data/processed/training_dataset.parquet")

print("Shape:", df.shape)
print("\nTarget distribution:")
print(df["target"].value_counts())

print("\nClientes com target=1:")
print(df[df["target"]==1]["customer_id"].nunique())

print("\nProdutos com mais target=1:")
print(df[df["target"]==1]["product_name"].value_counts().head(10))

print("\nNulls por coluna:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```
